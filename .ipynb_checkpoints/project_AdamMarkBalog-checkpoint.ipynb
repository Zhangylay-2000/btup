```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
import math
#used packages
```

```python

def datachoose(year=2020,states=None):
    """
    year: from 1996 to year the analyzed period
    states: list, names of locations
    """
    url1 = 'https://files.zillowstatic.com/research/public_v2/zhvi/Metro_zhvi_uc_sfrcondo_tier_0.67_1.0_sm_sa_mon.csv'
    url2="https://fred.stlouisfed.org/graph/fredgraph.csv?bgcolor=%23e1e9f0&chart_type=line&drp=0&fo=open%20sans&graph_bgcolor=%23ffffff&height=450&mode=fred&recession_bars=on&txtcolor=%23444444&ts=12&tts=12&width=1168&nt=0&thu=0&trc=0&show_legend=yes&show_axis_titles=yes&show_tooltip=yes&id=DGS3MO&scale=left&cosd=1996-01-31&coed=2021-03-23&line_color=%234572a7&link_values=false&line_style=solid&mark_type=none&mw=3&lw=2&ost=-99999&oet=99999&mma=0&fml=a&fq=Daily&fam=avg&fgst=lin&fgsnd=2020-02-01&line_index=1&transformation=lin&vintage_date=2021-03-25&revision_date=2021-03-25&nd=1981-09-01"
    homes = pd.read_csv(url1, sep=',')
    treasury = pd.read_csv(url2, sep=',').T
    date=[]
    for i in treasury.iloc[0]:
        date.append(datetime.datetime.strptime(i, '%Y-%m-%d'))
    treasury.columns=date
    treasury=treasury.drop('DATE')
    treasury=treasury[[i for i in treasury.columns if i.year<=year ]]
    #downloading data, changing date string to datetype
    date=[]
    for i in range(len(treasury.iloc[0])):
        if treasury.iloc[0][i] ==".":
            treasury.iloc[0][i]=0
        elif treasury.iloc[0][i] ==float("NaN") or treasury.iloc[0][i] ==float("Inf"):
            treasury.iloc[0][i]=0
    for i in homes.columns[5:]:
        date.append(datetime.datetime.strptime(i, '%Y-%m-%d'))
    homes=homes.set_index('RegionName')
    homes=homes.drop( list(homes.columns[:4]), axis=1)
    homes.columns=date
    homes=homes[[i for i in homes.columns if i.year<=year or (i.year==year+1 and i.month<=2)]]
    return homes.loc[states], treasury #returns 2 dataframes: home values in the appropiate locations, and the treasury rate over the period

places=['Atlantic City, NJ', 'New London, CT', 'New York, NY', 'Washington, DC', 'Santa Cruz, CA', 'Cedar Rapids, IA'] #just a few locations
```

```python
def diffpercentannualized(data):
    """
    data: pd.Series
    """
    percentages=[1] #starting percentage
    for i in range(len(data.to_list())-1):
        percentages.append((((data.to_list()[i+1])/data.to_list()[i])))
    #calculating the percentage growths month by month
    threemonth=[]
    for i in range(int(len(percentages)-2)):
        threemonth.append((percentages[i]*percentages[i+1]*percentages[i+2]))
    #calculating the 3 month returns of data
    return ((np.array(threemonth)**4)-np.ones(len(threemonth)))*100 #returns an array of the annualized percentage growths calculated from 3 month returns
```

```python
def finaltable(year, states):
    """
    year: from 1996 to year the analyzed period
    states: list, names of locations
    """
    homes, treasury = datachoose(year=year, states=states) #downloading and data manipulation
    treasury.loc["annualized"]=np.array(treasury.loc["DGS3MO"].to_numpy().astype(np.float)) #changing trasury rates to float
    treasury=treasury.drop("DGS3MO") #dropping the string row
    annualizedhome=[]
    for i in states:
        annualizedhome.append(diffpercentannualized(homes.loc[i]))
    #collecting annualized percenteges of home values by locations
    finalhome=pd.DataFrame(columns=homes.columns.to_list()[:-2]) #final table construction which contains the appropiate data about the annualized home returns
    for i in range(len(states)):
        finalhome.loc[states[i]]=annualizedhome[i]
    finaltreasure=pd.DataFrame(columns=treasury.columns)
    finaltreasure.loc["Treasury% annualized"]=treasury.loc['annualized']
    finaltreasure=finaltreasure[[i for i in finaltreasure.columns if finaltreasure[i][0]!=0]] #final table construction which contains the appropiate type of data about the treasury rates, excluding 0-s
    return finalhome, finaltreasure
```

```python
def graphs(year, states):
    """
    year: from 1996 to year the analyzed period
    states: list, names of locations
    """
    tablehome, tabletrs=finaltable(year, states) #final form of tables
    for i in states: #plotting
        fig=plt.figure(figsize=(20,10)) #size
        ax=fig.add_subplot(111, label="1")
        ax2=fig.add_subplot(111, label="2", frame_on=False) #overlapping figures
        ax.plot(tablehome.loc[i].index, tablehome.loc[i], color="C0",label=i)
        ax.legend()
        if math.isnan(min(min(tablehome.loc[i].to_list()),min(tabletrs.iloc[0].to_list()))-4) is False:
            ax.set_ylim((min(min(tablehome.loc[i].to_list()),min(tabletrs.iloc[0].to_list()))-4,max(max(tablehome.loc[i].to_list()),max(tabletrs.iloc[0].to_list()))+4))
        else: ax.set_ylim((-30,30))
        """
        setting y lim to appropiate value
        if statement: sometimes min/max gives nan value at 1 or 2 locations, in this case the default limits: (-30,30)
        same with the 2nd plot
        """
        ax2.plot(list(range(len(tabletrs.iloc[0]))),tabletrs.iloc[0], color="C1",label="US Treasury Annualized return")
        ax2.legend(loc=2)
        ax2.yaxis.set_label_position('right') 
        if math.isnan(min(min(tablehome.loc[i].to_list()),min(tabletrs.iloc[0].to_list()))-4) is False:
            ax2.set_ylim((min(min(tablehome.loc[i].to_list()),min(tabletrs.iloc[0].to_list()))-4,max(max(tablehome.loc[i].to_list()),max(tabletrs.iloc[0].to_list()))+4))
        else: ax2.set_ylim((-30,30))
        ax2.xaxis.set_visible(False)
        plt.title("3-month USD Treasury Constant Maturity Rate Vs. Annualized return of home values over 3 months in "+i,fontdict={'fontsize': 15})
        plt.axhline(y=0, color='black', linestyle='-') #showing y=0 axis
        plt.fill_between(np.linspace(0,len(tabletrs.iloc[0]),3),min(min(tablehome.loc[i].to_list()),min(tabletrs.iloc[0].to_list()))-4,color="maroon",alpha=0.4) #colorize the area under that
        plt.show()
        #returns nothing, just draws
graphs(2020, places)
graphs(2010, places)
```
