# Wiener process

1. Consider the coin flipping process with $\,n\,$ steps
2. Rescale time by $\,n^{\,-1}\,$
3. Rescale the cumulated sum of $\,+1\,$ and $\,-1\,$ vaules by $\,n^{\,-1\,\mathit{/}\,2}\,$
4. Take the $\,n\to\infty\,$ limit

<span style='background-color:yellow;color:red;font-weight:bold'>+ Illustration of drift and random walk with paths, line and +/- sqrt guides</span>

```python
import matplotlib.pyplot as plt
import numpy as np
from numbers import Number

N_LIST = 100, 1000, 10000
COLOR_LIST = 'darkgreen', 'blue', 'red'
LINE_WIDTH_LIST = 0.4, 0.2, 0.1
REALIZATIONS = 20


def plot_wiener_approx(n_list=None, color_list=None, line_width_list=None, realizations=None):
    """
    Plot discrete approximations of the Wiener process
    """

    # --- checking and setting values ---
    assert isinstance(n_list, (list, tuple)), 'n_list has to be a list or tuple'
    assert isinstance(n_list[0], Number), 'n_list must contain numbers'
    assert isinstance(color_list, (list, tuple)), 'color_list has to be a list or tuple'
    assert isinstance(realizations, int), '"realizations" must be an integer'
    assert isinstance(line_width_list, (list, tuple)), 'line_width_list has to be a list or tuple'
    
    my_fontsize = 16
    figure_size = (13, 8)
    labelpad = 20
    
    # --- plotting ---
    plt.figure(figsize=figure_size)
    for color, n, line_width in zip(color_list, n_list, line_width_list):
        for i_realization in range(realizations):
            random_steps = [1.0 if i == 1 else -1.0 for i in np.random.randint(0, high=2, size=n)]
            rescaled_cumulated_random_steps = [x / np.sqrt(n) for x in np.cumsum(random_steps)]
            plt.plot(np.linspace(0, 1.0, n, endpoint=False), rescaled_cumulated_random_steps, linestyle='-',
                     label='{} steps'.format(n) if 0 == i_realization else None, color=color, lw=line_width)

    plt.legend(bbox_to_anchor=(1.02, 1.0), loc=2, borderaxespad=0., fontsize=my_fontsize)
    plt.show()


plot_wiener_approx(n_list=N_LIST, color_list=COLOR_LIST, line_width_list= LINE_WIDTH_LIST, realizations=REALIZATIONS)
```
